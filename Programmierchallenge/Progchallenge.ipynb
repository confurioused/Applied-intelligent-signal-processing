```python
"""
Aufgabe: Trainieren Sie einen Klassifikator, der optimal eine Diabetesdiagnose prognostiziert.

Eingangsdaten: training_data.csv

Abzugeben sind zwei Skripte/Funktionen:
trainClassifier (Trainiert mithilfe von training_data.csv einen Klassifikator)
predictDiagnosis (Klassifiziert Eingangsdaten, die das gleiche Format wie training_data.csv besitzen)

Erforderliche Doku: Begründung der Auswahl des Klassifikators und geeignete Evaluation."""

"""
Bei einer linearen Trennung der beiden Klassen würde es aufgrund des 8 dimensionalen Parameterraumes viele unbestimmte Bereiche
geben, bei denen nicht klar definiert werden kann zu welcher Klasse die Merkmalsvektoren gehören. 
Viele der Methoden Basieren auf einer Normalverteilung der beiden Klassen. Da die Merkmale jedoch selber Normalverteilt sind,
ist hier nicht zu erwarten, dass die beiden Klassen je Merkmal eine Normalverteilung gut approximieren. 
Im folgenden Programm wird ein k Nächste Nachbarn Algorithmus angesetzt.
dieser Algorithmus zählt zu den Parameterfreien. Die Funktion find_best_KNN sucht die besten Hyperparameter 
(k, Distanz und ggf. p)

Die Genauigkeit des gefundenen Netzes liegt in der Nähe von 80%. Allerdings werden die meisten Vorhersagen auf 1 gesetzt.
Aufgrund der hohen Dimensionalität lässt sich die Trennfunktion schlecht grafisch überprüfen.

"""
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import math
from sklearn.neighbors import KNeighborsClassifier
data = pd.read_csv("C:\\Users\\aaron\\OneDrive\\Dokumente\\Uni\\Angewandte intelligente Signalverarbeitung\\Aarons Praktikum\\Programmierchallenge\\training_data.csv")

X = data.drop(columns=['Outcome'])
y= data["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

stdScaler = StandardScaler()
stdScaler.fit(X_train)
X_train = stdScaler.transform(X_train)
X_test = stdScaler.transform(X_test)
```

```python
def find_best_KNN(X_train, X_test, y_train, y_test, k):
    realdist = ["euclidean","manhattan","chebyshev","minkowski","mahalanobis"]
    np.array(X_train)
    covariance = np.cov(X_train)
    #best = (score,dist,param)
    best = (0,"euclidean",None)
    for dist in realdist:
        if(dist =="minkowski"):
            
            for p in range(1,6,1):
                knn = KNeighborsClassifier(n_neighbors=k, metric=dist, metric_params={"p" : p})
                knn.fit(X_train, y_train)
                score = knn.score(X_test, y_test)
                if(best[0]< score):
                    best = (score,dist,p)
                

        elif(dist =="mahalanobis"):
            
            knn = KNeighborsClassifier(n_neighbors=k, metric=dist,  metric_params={'V': np.cov(X_train.T)})
            knn.fit(X_train, y_train)
            
            score = knn.score(X_test, y_test)
            if(best[0]< score):
                    best = (score,dist,None)
        else:
            
            knn = KNeighborsClassifier(n_neighbors=k, metric=dist)
            knn.fit(X_train, y_train)
            
            score = knn.score(X_test, y_test)
            if(best[0]< score):
                    best = (score,dist,None)
    return best
```

```python

#rule of thumb for neighbors
#k = math.ceil((data.count().unique()[0]/2)**0.5) = 19
best_scores = {}
for k in range(2,100,1):
    best_scores[k] = find_best_KNN(X_train, X_test, y_train, y_test,k)
    
k_best = max(best_scores, key=best_scores.get)
final_params = best_scores[k_best]
print("best hyperparameters found: ","distance: ", best_scores[k_best][1],"parameters: ", best_scores[k_best][2],"score: ", best_scores[k_best][0], "k :",k_best) 
```

```python
print("best hyperparameters found: ","distance -", best_scores[k_best][1],"; parameters - ", best_scores[k_best][2],"; score - ", best_scores[k_best][0], "; k -",k_best)
```

```python
def trainClassifier(X,y,final_params = (0,"euclidean",None)):
    stdScaler = StandardScaler()
    stdScaler.fit(X)
    X = stdScaler.transform(X)
    if(final_params[1] == "minkowski"):
        final_knn = KNeighborsClassifier(n_neighbors=k_best, metric=final_params[1],  metric_params={"p" : final_params[2]})
    else:
        final_knn = KNeighborsClassifier(n_neighbors=k_best, metric=final_params[1])
    final_knn.fit(X, y)
    return stdScaler,final_knn
```

```python
def predictDiagnosis(scaler,model,x):
    x= x.drop(columns=['Outcome'])
    scaler.transform(x)
    return model.predict(x)
    
    
```

```python
[scaler,model] = trainClassifier(X,y,final_params)
```
